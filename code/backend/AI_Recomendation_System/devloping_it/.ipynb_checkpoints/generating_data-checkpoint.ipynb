```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler, MultiLabelBinarizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Constants (same as previous script)
SERVICE_CATEGORIES = [
    "Plumbing", "Electrical Work", "Painting and Decoration", "Masonry",
    "Residential Cleaning", "Commercial Cleaning", "Post-Construction Cleaning",
    "Taxi Services", "Furniture Moving", "Local Delivery Services",
    "Car Repair and Maintenance", "Tire Services", "Car Wash",
    "Traditional Algerian Catering", "Event Catering",
    "Babysitting", "Elderly Companionship Services",
    "Language Lessons", "Quranic Studies", "Academic Tutoring",
    "Hairdressing", "Makeup Services", "Hammam and Spa Services"
]

def haversine_distance(lat1, lon1, lat2, lon2):
    """Calculate the great circle distance between two points."""
    R = 6371  # Earth radius in kilometers
    
    # Convert latitude and longitude to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    distance = R * c
    
    return distance

def encode_categorical_features(series, unique_categories):
    """One-hot encode categorical features."""
    return pd.get_dummies(series, prefix='', prefix_sep='', columns=unique_categories)

def generate_city_features(city_users):
    """Generate advanced features for clustering."""
    # One-hot encode gender and service categories
    gender_encoded = encode_categorical_features(city_users['gender'], ['Male', 'Female'])
    
    # Multi-label encoding for service categories
    mlb = MultiLabelBinarizer(classes=SERVICE_CATEGORIES)
    service_interest_encoded = pd.DataFrame(
        mlb.fit_transform(city_users['service_categories_interest']), 
        columns=SERVICE_CATEGORIES
    )
    
    # Combine features with weighted importance
    features = pd.concat([
        city_users[['location_x', 'location_y']] * 2,  # Emphasize location
        city_users[['age']],
        gender_encoded,
        service_interest_encoded * 1.5  # Higher weight for service interests
    ], axis=1)
    
    # Standardize features
    scaler = StandardScaler()
    return scaler.fit_transform(features), scaler, mlb

def train_city_clusters(user_df, max_clusters=10, min_samples_per_cluster=5):
    """Advanced city-based clustering with adaptive cluster determination."""
    city_clusters = {}
    all_city_scalers = {}
    all_city_encoders = {}
    
    for city in user_df['city'].unique():
        city_users = user_df[user_df['city'] == city].reset_index(drop=True)
        
        # Dynamically determine number of clusters
        num_users = len(city_users)
        potential_clusters = min(max_clusters, num_users // min_samples_per_cluster)
        
        # Find optimal number of clusters using silhouette score
        best_n_clusters = 2
        best_score = -1
        
        for k in range(2, potential_clusters + 1):
            features, _, _ = generate_city_features(city_users)
            
            try:
                kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
                cluster_labels = kmeans.fit_predict(features)
                
                # Validate cluster quality
                if len(set(cluster_labels)) == k:
                    score = silhouette_score(features, cluster_labels)
                    if score > best_score:
                        best_score = score
                        best_n_clusters = k
            except Exception as e:
                print(f"Clustering error for {city} with {k} clusters: {e}")
        
        # Final clustering
        features, scaler, encoder = generate_city_features(city_users)
        kmeans = KMeans(n_clusters=best_n_clusters, random_state=42, n_init=10)
        city_users['cluster'] = kmeans.fit_predict(features)
        
        city_clusters[city] = city_users
        all_city_scalers[city] = scaler
        all_city_encoders[city] = encoder
    
    return city_clusters, all_city_scalers, all_city_encoders

def generate_service_recommendations(city_clusters, service_df, top_k=20):
    """Generate targeted service recommendations for each city and cluster."""
    all_recommendations = {}
    
    for city, clustered_users in city_clusters.items():
        city_recommendations = {}
        
        for cluster in clustered_users['cluster'].unique():
            cluster_users = clustered_users[clustered_users['cluster'] == cluster]
            
            # Aggregate cluster service interests
            cluster_interests = cluster_users['service_categories_interest'].explode()
            
            # Recommendation scoring
            recommended_services = service_df[
                service_df['service_category'].isin(cluster_interests)
            ].copy()
            
            # Complex ranking algorithm
            recommended_services['recommendation_score'] = (
                recommended_services['review_avg'] * 0.7 +
                recommended_services['click_count'].rank(method='min') * 0.15 +
                recommended_services['review_count'].rank(method='min') * 0.15
            )
            
            top_recommendations = recommended_services.nlargest(top_k, 'recommendation_score')
            city_recommendations[cluster] = top_recommendations
        
        all_recommendations[city] = city_recommendations
    
    return all_recommendations


# Run clustering
city_cluster_results, all_city_scalers, all_city_encoders = train_city_clusters(user_df)

# Generate recommendations
service_recommendations = generate_service_recommendations(city_cluster_results, service_df)

# Save models and recommendations
def save_recommendation_models(city_clusters, scalers, encoders, recommendations, filename="recommendation_models.pkl"):
    joblib.dump({
        'city_clusters': city_clusters,
        'scalers': scalers,
        'encoders': encoders,
        'recommendations': recommendations
    }, filename)
    print(f"Models and recommendations saved to {filename}")

save_recommendation_models(
    city_cluster_results, 
    all_city_scalers, 
    all_city_encoders, 
    service_recommendations
)

# Display some example results
for city, city_clusters in service_recommendations.items():
    print(f"\nRecommendations for {city}:")
    for cluster, recommendations in city_clusters.items():
        print(f"  Cluster {cluster} Top Services:")
        print(recommendations[['service_category', 'review_avg', 'click_count']].head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score, adjusted_rand_score

def evaluate_clustering_quality(city_clusters):
    """
    Comprehensive clustering quality evaluation
    
    Parameters:
    city_clusters (dict): Clustered user data for each city
    
    Returns:
    dict: Clustering quality metrics
    """
    clustering_metrics = {}
    
    for city, clustered_users in city_clusters.items():
        # Extract clustering features
        features, _, _ = generate_city_features(clustered_users)
        clusters = clustered_users['cluster']
        
        city_metrics = {
            'num_clusters': len(np.unique(clusters)),
            'num_users': len(clustered_users),
            'silhouette_score': silhouette_score(features, clusters),
            'cluster_distribution': dict(clusters.value_counts())
        }
        
        # Calculate inter-cluster diversity
        cluster_interests = {}
        for cluster in np.unique(clusters):
            cluster_users = clustered_users[clustered_users['cluster'] == cluster]
            interests = cluster_users['service_categories_interest'].explode()
            cluster_interests[cluster] = list(interests.value_counts().head())
        
        city_metrics['cluster_top_interests'] = cluster_interests
        clustering_metrics[city] = city_metrics
    
    return clustering_metrics

def visualize_clustering_results(city_clusters):
    """
    Create visualizations to understand clustering
    
    Parameters:
    city_clusters (dict): Clustered user data for each city
    """
    plt.figure(figsize=(15, 10))
    
    for i, (city, clustered_users) in enumerate(city_clusters.items(), 1):
        plt.subplot(2, 3, i)
        
        # Scatter plot of users with cluster colors
        plt.scatter(
            clustered_users['location_x'], 
            clustered_users['location_y'], 
            c=clustered_users['cluster'], 
            cmap='viridis'
        )
        plt.title(f'User Clusters in {city}')
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')
    
    plt.tight_layout()
    plt.savefig('clustering_visualization.png')
    plt.close()

def evaluate_recommendation_quality(service_recommendations, service_df, user_df):
    """
    Evaluate recommendation quality
    
    Parameters:
    service_recommendations (dict): Generated recommendations
    service_df (DataFrame): Service dataset
    user_df (DataFrame): User dataset
    
    Returns:
    dict: Recommendation quality metrics
    """
    recommendation_metrics = {}
    
    for city, city_clusters in service_recommendations.items():
        city_metrics = {
            'total_clusters': len(city_clusters),
            'cluster_recommendation_diversity': {},
            'recommendation_coverage': {}
        }
        
        for cluster, recommendations in city_clusters.items():
            # Diversity of recommended services
            recommended_categories = recommendations['service_category'].unique()
            
            # Coverage of user interests
            city_cluster_users = user_df[
                (user_df['city'] == city) & 
                (user_df['cluster'] == cluster)
            ]
            user_interests = city_cluster_users['service_categories_interest'].explode()
            
            coverage_score = len(set(recommended_categories) & set(user_interests)) / len(set(user_interests))
            
            city_metrics['cluster_recommendation_diversity'][cluster] = {
                'unique_categories': len(recommended_categories),
                'top_recommended_categories': list(recommendations['service_category'].value_counts().head())
            }
            
            city_metrics['recommendation_coverage'][cluster] = coverage_score
        
        recommendation_metrics[city] = city_metrics
    
    return recommendation_metrics

def comprehensive_model_evaluation(
    city_clusters, 
    service_recommendations, 
    service_df, 
    user_df
):
    """
    Perform comprehensive model evaluation
    
    Returns:
    dict: Detailed model evaluation report
    """
    # Clustering Quality Evaluation
    clustering_metrics = evaluate_clustering_quality(city_clusters)
    
    # Visualization of Clustering
    visualize_clustering_results(city_clusters)
    
    # Recommendation Quality Evaluation
    recommendation_metrics = evaluate_recommendation_quality(
        service_recommendations, 
        service_df, 
        user_df
    )
    
    # Combine metrics
    evaluation_report = {
        'clustering_metrics': clustering_metrics,
        'recommendation_metrics': recommendation_metrics
    }
    
    # Generate detailed report
    with open('model_evaluation_report.txt', 'w') as f:
        f.write("Recommendation System Evaluation Report\n")
        f.write("=" * 40 + "\n\n")
        
        f.write("Clustering Metrics:\n")
        for city, metrics in clustering_metrics.items():
            f.write(f"\nCity: {city}\n")
            f.write(f"  Number of Clusters: {metrics['num_clusters']}\n")
            f.write(f"  Total Users: {metrics['num_users']}\n")
            f.write(f"  Silhouette Score: {metrics['silhouette_score']:.4f}\n")
            f.write("  Cluster Distribution:\n")
            for cluster, count in metrics['cluster_distribution'].items():
                f.write(f"    Cluster {cluster}: {count} users\n")
            
            f.write("  Top Cluster Interests:\n")
            for cluster, interests in metrics['cluster_top_interests'].items():
                f.write(f"    Cluster {cluster}: {interests}\n")
        
        f.write("\nRecommendation Metrics:\n")
        for city, metrics in recommendation_metrics.items():
            f.write(f"\nCity: {city}\n")
            f.write(f"  Total Clusters: {metrics['total_clusters']}\n")
            
            f.write("  Cluster Recommendation Diversity:\n")
            for cluster, div_metrics in metrics['cluster_recommendation_diversity'].items():
                f.write(f"    Cluster {cluster}:\n")
                f.write(f"      Unique Categories: {div_metrics['unique_categories']}\n")
                f.write(f"      Top Recommended: {div_metrics['top_recommended_categories']}\n")
            
            f.write("  Recommendation Coverage:\n")
            for cluster, coverage in metrics['recommendation_coverage'].items():
                f.write(f"    Cluster {cluster}: {coverage:.2%}\n")
    
    # Optional: Create visualization of recommendation coverage
    plt.figure(figsize=(10, 6))
    coverages = [
        metrics['recommendation_coverage'] 
        for metrics in recommendation_metrics.values()
    ]
    sns.boxplot(data=[list(coverage.values()) for coverage in coverages])
    plt.title('Recommendation Coverage Across Cities')
    plt.xlabel('Cities')
    plt.ylabel('Coverage Score')
    plt.savefig('recommendation_coverage.png')
    plt.close()
    
    return evaluation_report

# Example usage (add to your main script)
model_evaluation = comprehensive_model_evaluation(
    city_cluster_results, 
    service_recommendations, 
    service_df, 
    user_df
)

# Print key highlights
print("Model Evaluation Completed. Check 'model_evaluation_report.txt' for details.")
print("Visualizations saved: 'clustering_visualization.png', 'recommendation_coverage.png'")
```
